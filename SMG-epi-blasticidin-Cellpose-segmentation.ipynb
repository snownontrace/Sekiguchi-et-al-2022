```python
import numpy as np
import time, os, sys, glob
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
mpl.rcParams['figure.dpi'] = 300
from cellpose import models, io
```

```python
# data_folder = '../data/220601-X-188-blast-test/220601-X-188-add-blast-40x-output'
data_folder = '../data/220601-X-188-blast-test/220601-X-188-no-blast-40x-output'
files = glob.glob(os.path.join(data_folder, '*.tif'))

# view the DAPI channel of the last image
img = io.imread(files[-1])
plt.figure(figsize=(2,2))
plt.imshow(img[0])
plt.axis('off')
plt.show()
```

```python
# DEFINE CELLPOSE MODEL
# model_type='cyto' or 'nuclei'
model = models.Cellpose(gpu=False, model_type='cyto')

# define CHANNELS to run segementation on
# channels = [cytoplasm, nucleus]
# our tif images contain 3 channels that are DAPI, GFP and NLS-mScarlet
CHANNELS = [1, 0]

# if diameter is set to None, the size of the cells is estimated on a per image basis
# you can set the average cell `diameter` in pixels yourself (recommended) 
# diameter can be a list or a single number for all images
Diameter = 50

for filename in files:
    img = io.imread(filename)
    masks, flows, styles, diams = model.eval(img, diameter=Diameter, channels=CHANNELS)

    # save results so you can load in gui
    io.masks_flows_to_seg(img, masks, flows, diams, filename, CHANNELS)

    # save results as png
    io.save_to_png(img, masks, flows, filename)
```
